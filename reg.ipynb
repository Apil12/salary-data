```python
new_df['Gender'].replace("Male",1,inplace=True)
new_df['Gender'].replace("Female", 0, inplace=True)                      
```

```python
new_df['Education Level'].replace("Bachelor's", 0, inplace=True)
new_df['Education Level'].replace("Master's", 1, inplace=True)
new_df['Education Level'].replace("PhD", 2, inplace=True)
```

```python
new_df
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data=train_test_split(new_df,train_size=0.7,test_size=0.3,random_state=1)
train_data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler
```

```python
vare=['Age','Years of Experience','Salary']
train_data[vare] = scaler.fit_transform(train_data[vare])
train_data[vare]
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(new_df.corr(), annot = True,)
plt.show()
```

```python
y_train = train_data.pop('Salary')
X_train = train_data
(X_train)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

#rfe = RFE(lm, 10)
rfe = RFE(lm, n_features_to_select=12)
#running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train_rfe = X_train[col]
X_train_rfe
```

```python
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
X_train_rfe
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

```python
print(lm.summary())
```

```python
X_train_new1 = X_train_rfe.drop(["Age"], axis = 1)
X_train_new1
```

```python
lm = sm.OLS(y_train,X_train_new1).fit()   # Running the linear model
```

```python
print(lm.summary())
```

```python
y_train_price = lm.predict(X_train_new1)
y_train_price
```

# steps for linear regression

```python
# At first, file is readed and dataset is converted into dataframe.
# null values are removed from dataset EDA has been performed to analysis of data.
# Dimension reduction has been done to convert categorical into numerical data..
# dataset is converted into machine learning train and test data.
# minmax scalar algorithm has been implemented to normalaize the dataset (to convert high value of data into lower ones..)
# heatmap has been plotted for dataframe to see co-relation  matrix it shows that which values are same means
# columns are highly corelated and which values need to be removed..
# Although we can removed those manually but we use new technique called RFE to removed those unwanted columns.
# After that ols technique is used to find coeffiecent for the equations it shows summary of the whole dataset too..
# after that we need to test the model and find out y pred also using r2 score method accuracy of model is calculated..
```
